```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Data

Read the data from our .csv file.

```python
df = pd.read_csv('load_per_country.csv', parse_dates=['DateTime'])

print(df.shape)
```

```python
select = df[['DateTime','PL','GR']]
print(select)
```

### Graphing the Data - Raw

```python
x = 'DateTime'
y1 = 'PL'
y2 = 'GR'

df.plot(x, y=y1, color='r',  label='pl')
df.plot(x, y=y2, color = 'g',  label = 'gr')
plt.legend()
plt.show()
```

### Graphing the Data - Resampled

The above plots can look a bit crowded as it is drawing hourly values. To resample the data, use the below method. Try replacing the `freq` parameter with "Y" for year, etc.

```python
monthly = df.groupby(pd.Grouper(key='DateTime', axis=0, freq='M')).sum()
monthly.plot(y='PL', color = 'r', label = 'pl')
monthly.plot(y='GR', color = 'g', label='gr')
plt.show()
```
