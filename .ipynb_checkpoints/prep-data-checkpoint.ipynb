```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Data

Read the data from our .csv file.

```python
df = pd.read_csv('load_per_country.csv', parse_dates=['DateTime'], index_col = ['DateTime'])

print(df.shape)
```

```python
select = df[['PL','DE']]
print(select)
```

### Graphing the Data - Raw

```python
select.plot()
plt.legend()
plt.title("Load over Time (MW)")
plt.show()
```

### Graphing the Data - Resampled

The above plots can look a bit crowded as it is drawing hourly values. To resample the data, use the below method. Try replacing the `freq` parameter with "Y" for year, "24H" for day, etc.

```python
resampled = select.groupby(pd.Grouper(level='DateTime', axis=0, freq='M')).sum()
resampled.plot()
plt.title("Load over Time (MW), Monthly Resample")
```
